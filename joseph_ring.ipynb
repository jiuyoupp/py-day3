```python
def josephus1(num, step):
    output = []
    p=0
    people = list(range(1, num+1))
   
    while len(people) > 1:
        p = (p+(step-1)) % len(people)
        temp = people.pop(p)
        output.append(temp)
    output.append(people[0])

    return output


try:
    num = int(input("输入总人数"))
    interval = int(input("输入间隔"))
    outseq = josephus1(num, interval)
except ValueError:
    print('输入错误，输入的不是数字')
else:
    print(outseq)  
```

```python
class Node(object):
    def __init__(self, item):
        self.elem = item
        self.next = None


class SingleCycleLinkList(object):
    """单向循环链表"""
    def __init__(self):
        self.head = None

    def is_empty(self):
        # 链表是否为空
        return self.head == None

    def travel_end(self):
        # 遍历到尾节点
        cur = self.head
        while cur.next != self.head:
            cur = cur.next
        return cur

    def append(self, item):
        # 尾插法
        node = Node(item)
        # 节点为空
        if self.is_empty():
            self.head = node
            node.next = node
        else:
            cur = self.travel_end()
            cur.next = node
            node.next = self.head

    def remove(self, item):
        # 删除结点
        if self.is_empty():
            return
        cur = self.head
        pre = None

        while cur.next != self.head:
            if cur.elem == item:
                # 删除位置在头节点且多于一个元素的情况
                if cur == self.head:
                    # 先找尾节点
                    rear = self.travel_end()
                    rear.next = cur.next
                    self.head = cur.next
                else:
                    # 中间结点和尾结点的情况
                    pre.next = cur.next
                return
            else:
                pre = cur
                cur = cur.next

        # 退出循环指向尾结点
        if cur.elem == item:
            #  一个元素的情况删掉后变为空节点  防止pre为none导致出错
            if cur == self.head:
                self.head = None
            else:
                # 尾节点元素的情况
                pre.next = self.head

    def judgement(self, interval):
        cur = self.head
        count = 1
        output = []
        while cur != cur.next:
            cur = cur.next
            count += 1
            if count == interval:
                output.append(cur.elem)
                self.remove(cur.elem)
                count = 0
        output.append(cur.elem)

        return output


def joseph(num, interval):
    jos = SingleCycleLinkList()
    for i in range(1, num+1):
        jos.append(i)
    out = jos.judgement(interval)
    return out


try:
    num = int(input("输入总人数"))
    interval = int(input("输入间隔"))
    outseq = joseph(num, interval)
except ValueError:
    print('输入错误，输入的不是数字')
else:
    print(outseq)
```
